## 4-1-1. [공공자전거_대여이력_정보] 데이터 불러오기

```python
trip19 = pd.read_csv("/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/rental_data_2019.csv", encoding='cp949')
trip20 = pd.read_csv("/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/rental_data_2020.csv", encoding='cp949')
trip21 = pd.read_csv("/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/rental_data_2021.csv", encoding='cp949')
```

```python
trip = pd.concat([trip19,trip20,trip21])
```

```python
trip.info()
```

```python
trip.head()
```

## 4-1-2. 자치구 데이터와 결합 "trip_area"

```python
trip_area = pd.merge(trip,area, how='outer',on='rental_loc_no')
```

```python
trip_area
```

```python
trip_area['district'].unique() # 총 25개 자치구
```

## 4-1-3. 자치구별 총 이용건수 순위

```python
trip_area['district'].value_counts() #자치구별 이용건수
```

## 4-1-4. 자치구별 그룹 생성

```python
trip_grouped = trip_area.groupby('district')
```

```python
for key, group in trip_grouped:
    print("* key", key)
    print("* count", len(group))
    print(group.head())
    print('\n')
```

## 4-1-5. 자치구별 평균 이용시간, 평균 이용거리
