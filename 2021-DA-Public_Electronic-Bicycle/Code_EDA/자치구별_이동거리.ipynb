# 2. [공공자전거_대여이력_정보] 데이터 불러오기

```python
trip19 = pd.read_csv("/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/rental_data_2019.csv", encoding='cp949')
trip20 = pd.read_csv("/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/rental_data_2020.csv", encoding='cp949')
```

# 3. 자치구 데이터와 결합 "trip_area"

```python
trip19_area = pd.merge(trip19,area, how='outer',on='rental_loc_no')
trip20_area = pd.merge(trip20,area, how='outer',on='rental_loc_no')
```

```python
trip19_area.head()
```

# 4. 자치구별 총 이용건수

```python
trip19_area['district'].value_counts() #자치구별 이용건수
```

```python
trip20_area['district'].value_counts() #자치구별 이용건수
```

# 5. 자치구별 총 이용거리

```python
trip19_grouped = trip19_area.groupby('district')
trip20_grouped = trip20_area.groupby('district')
```

```python
trip19_grouped = trip19_grouped[['district', 'rental_time', 'rental_distance']]
trip20_grouped = trip20_grouped[['district', 'rental_time', 'rental_distance']]
```

```python
pd.options.display.float_format = '{:.5f}'.format
trip19_grouped_sum = trip19_grouped.sum() # 자치구별 총 이용시간(rental_time), 이용거리(rental_distance)
trip19_grouped_sum
```

```python
pd.options.display.float_format = '{:.5f}'.format
trip20_grouped_sum = trip20_grouped.sum() # 자치구별 총 이용시간(rental_time), 이용거리(rental_distance)
trip20_grouped_sum
```

```python
trip19_grouped_sum.to_csv('/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/trip19_distance_sum.csv', encoding = 'cp949')
```

```python
trip20_grouped_sum.to_csv('/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/trip20_distance_sum.csv', encoding = 'cp949')
```

# 6. 자치구별 평균 이용거리

```python
trip19_grouped_mean = trip19_grouped.mean() # 자치구별 평균 이용시간(rental_time), 이용거리(rental_distance)
trip19_grouped_mean
```

```python
trip20_grouped_mean = trip20_grouped.mean() # 자치구별 평균 이용시간(rental_time), 이용거리(rental_distance)
trip20_grouped_mean
```

```python
trip20_grouped_mean.to_csv('/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/trip20_distance_mean.csv', encoding = 'cp949')
trip20_grouped_mean.to_csv('/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#4. 공공자전거_대여이력_정보/trip20_distance_mean.csv', encoding = 'cp949')
```
