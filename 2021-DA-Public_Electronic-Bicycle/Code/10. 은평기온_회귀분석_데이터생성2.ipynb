```python
df['location_name'] = df.location.map(lambda x: x.split('_')[1])
```

```python
location = pd.read_csv('/Volumes/GoogleDrive/공유 드라이브/빅분기C조/ju_notes/data/#1. 공공자전거_대여소_정보/area.csv', encoding='cp949')
```

```python
df_ep = df_location[df_location.district=='은평구']
df_gn = df_location[df_location.district=='강남구']
```

```python
def process(df):
    df = df.copy()
    df.rename(columns={'Unnamed: 0':'time'}, inplace=True)
    df['date'] = df.time.map(lambda x: x[:10])
    
    return df
```

```python
df_ep = process(df_ep)
```

```python
df_gn = process(df_gn)
```

```python
df_ep.head()
```

```python
ep_count = df_ep.pivot_table(index=['date', 'location_name'], values='usage_time', aggfunc='sum')
gn_count = df_gn.pivot_table(index=['date', 'location_name'], values='usage_time', aggfunc='sum')
```

```python
ep_count = ep_count.reset_index()
gn_count = gn_count.reset_index()
```

```python
ep_count.head()
```

```python
gn_count.head()
```

```python
# 계절 붙이기
from datetime import datetime


def get_season(dt):
    month = int(dt[5:7])
    if month in [6,7,8]:
        season = 'summer'
    elif month in [12,1,2]:
        season = 'winter'
    else:
        season = 'else'
    return season
    
def is_weekday(dt):
    
    obj = datetime.strptime(dt, '%Y-%m-%d')
    day = obj.weekday()
    if day in [5,6]:
        return 0
    else:
        return 1
```

```python
ep_count['season'] = ep_count.date.map(get_season)
ep_count['weekday'] = ep_count.date.map(is_weekday)
```

```python
gn_count['season'] = gn_count.date.map(get_season)
gn_count['weekday'] = gn_count.date.map(is_weekday)
```
